```python
#remove unwanted columns
df= data.drop(['Z_CostContact','Z_Revenue','Complain'],axis=1)
df
```

```python
#calculate age from year birth column
df['age'] = 2022 - df['Year_Birth']
```

```python
df['age'].plot.box()
```

```python
#remove the outliers from dataset where age > 100
df = df[(df["age"]<100)]
```

```python
#Verify outliers
df['age'].plot.box()
```

```python
#histogram of Income
plt.figure(figsize=(20,8))
sns.histplot(df['Income'], color="blue")
plt.show()
```

```python
#remove the outliers from income
df = df[df['Income'] < 600000]
```

```python
#calculate total spend on products
df['total_spend'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] 
+ df['MntSweetProducts'] + df['MntGoldProds'] + df['NumDealsPurchases'] + df['NumWebPurchases']
```

```python
df.columns
```

```python
#correlation
# Increase the size of the heatmap.

plt.figure(figsize=(20,20))  
corr = df.corr()
sns.heatmap(corr, vmin=-1, vmax=1, annot= True,cmap='coolwarm_r',annot_kws={"fontsize":12}, fmt=".2f")
plt.show()
#vmin, vmax â€” set the range of values that serve as the basis for the colormap
#when set to True, the correlation values become visible on the colored cells
```

```python
#seperate catagorical and numerical columns
Catagorical = ['Kidhome','Teenhome','Education','Marital_Status','Response',
               'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
Numericals = []
for col in df.columns:
    if not col in Catagorical:
        Numericals.append(col)

del Numericals[0:2]
del Numericals[1:2]
Numericals
```

```python
#bar plot for total spend and eduction

#plt.figure(figsize=(10,8))

for x in Catagorical:
    #fig, axes = plt.subplots(2, sharex=True, figsize=(20,15))
    #sort = df.sort_values('total_spend')[x]
    sns.barplot(x= df[x],y='total_spend',data=df, ci=None)
    plt.ylabel("total spend in US Dollars", size=15)
    #plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
    plt.show()
```
